```python
import pandas as pd

baseball_df = pd.read_csv('./Data/baseball_stats.csv')
print(baseball_df.head())
```

```python
# Calculating win percentage

import numpy as np

def calc_win_perc(wins, games_played):
    
    win_perc = wins / games_played
    
    return np.round(win_perc, 2)
```

```python
win_perc = calc_win_perc(50, 100)
print(win_perc)
```

```python
# Adding win percentage to DataFrame
# Iterating with .iloc
%%timeit
win_perc_list = []

for i in range(len(baseball_df)):
    row = baseball_df.iloc[i]
    wins = row['W']
    games_played = row['G']
    win_perc = calc_win_perc(wins, games_played)
    win_perc_list.append(win_perc)
baseball_df['WP'] = win_perc_list
```

```python
# Iterating with .iterrows()
# .iterrows()returns each DataFrame row as tuple of (index, pandas Series) pairs. 

%%timeit
win_perc_list = []

for i,row in baseball_df.iterrows():
    wins = row['W']
    games_played = row['G']
    
    win_perc = calc_win_perc(wins, games_played)
    
    win_perc_list.append(win_perc)
    
baseball_df['WP'] = win_perc_list
```

```python
%%timeit
import pandas as pd

data = {
  "firstname": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df = pd.DataFrame(data)

for row in df.itertuples():
      print(row)
        
# Iterating with .itertuples()
# Iterate the rows of the DataFrame, and print each row as a tuple:
```

```python
# .apply()
# Pandas.apply allow the users to pass a function and apply it on every 
# single value of the Pandas series.


%%timeit

def calc_run_diff(runs_scored, runs_allowed):
    run_diff = runs_scored - runs_allowed
    return run_diff

diff = baseball_df.apply(lambda row: calc_run_diff(row['RS'], row['RA']),
                                                  axis = 1)

baseball_df['RD'] = diff
```
