# String operations

```python
my_string = "tHis Is a niCe StriNg"
print(my_string.lower())
print(my_string.upper())
print(my_string.capitalize())
```

```python
# Splitting
# Split starts splitting at the left

my_string = "This string will be split"

my_string.split(sep=" ", maxsplit=2)
```

```python
# rsplit begins at the right of the string. If maxsplit is not specified 
# both methods behave in the same way. 

my_string.rsplit(sep=" ", maxsplit=2)
```

```python
# the string is split at the slash n sequence returning a list of two elements.

my_string = "This string will be split\nin two"
my_string.splitlines()
```

```python
# Joining

my_list = ["this", "would", "be", "a", "string"]
print(" ".join(my_list))
```

```python
# .strip()
#  The dot strip method will remove both leading and trailing characters.

my_string = " This string will be stripped\n"
my_string.strip()
```

```python
my_string = " This string will be stripped\n"
my_string.rstrip()
```

```python
my_string.lstrip()
```

```python
# Finding
# The dot find method returns the lowest index in the string where it can 
# find the substring, in this case, eight. 

my_string = "Where's Waldo?"
my_string.find("Waldo")
```
