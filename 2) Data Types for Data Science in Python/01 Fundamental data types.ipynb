```python
# Combine List, Append, Extend method
names = ['Ian', 'Victor', 'Mike', 'Sam']
names2 = ['Ivy', 'Cameron', 'Tom']
names_final = names + names2
names_final.append('Victoria')
names_final.extend(['Xi', 'Kyle'])
print(names_final.index('Mike'))
names_final.pop(4)
print(names_final)
```

```python
for name in sorted(names_final):
    print(name)
```

```python
### xxxxx ###

# Sorted Function
# The sorted() function returns a new list and does not affect the list 
# you passed into the function. 

import pandas as pd
baby_names = pd.read_csv('baby_names.csv')
records = []

for row in baby_names:
    records.append(row[3])

    for name in sorted(records):
        print(name)
```

```python
list1 = ['1', '2', '3', '4']
list2 = ['Ian', 'Sam', 'Tom', 'Ivy']
Tuple1_2 = list(zip(list1, list2))
print(Tuple1_2)
```

```python
num, name = Tuple1_2[0]
print(num)
print(name)
```

```python
# More unpacking in Loops
for num, name in Tuple1_2:
    print(num)
    print(name)
```

```python
### *** ###
for idx, item in enumerate(Tuple1_2):
    num, name = item
    print(idx, num, name)
```

```python
### *** ###
# Set are created from list 
# A Set is an unordered collection data type that is iterable, mutable 
# and has no duplicate elements.

fruits = ['apple', 'orange', 'kiwi', 'banana']
fruit_set = set(fruits)
print(fruit_set)
```

```python
fruit_set.add('leechey')
```

```python
fruit_set.add('mango')
```

```python
print(fruit_set)
```

```python
additional = ['pineapple', 'kiwi']
fruit_set.update(additional)
print(fruit_set)
```

```python
fruit_set.discard('kiwi')
print(fruit_set)
```

```python
fruit_set.pop()
```

```python
fruit_set.pop()
```

```python
print(fruit_set)
```

```python
# Finding all the data and the overlapping data between sets
fruit_1 = set(['apple', 'orange', 'mango', 'Kiwi'])
fruit_2 = set(['Kiwi',  'Leechey'])
fruit_1.union(fruit_2)
```

```python
fruit_1.intersection(fruit_2)
```

```python
fruit_1.difference(fruit_2)
```

```python
fruit_2.difference(fruit_1)
```
