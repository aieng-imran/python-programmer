```python
# Import relevant classes
# The clock that recorded the ride was 5 hours behind UTC. 
# Now if you print it, your datetime includes the UTC offset.

from datetime import datetime, timedelta, timezone

# US Eastern Standard time zone
ET = timezone(timedelta(hours=-5))
# Timezone-aware datetime
dt = datetime(2017, 12, 30, 15, 9, 3, tzinfo = ET)

print(dt)
```

```python
# .astimezone()

IST = timezone(timedelta(hours=5, minutes=30))
print(dt.astimezone(IST))
```

```python
# Adjusting timezone vs changing tzinfo
# Here we've set the tzinfo to be timezone.utc, a convenient object 
# with zero UTC offset. 

print(dt)

print(dt.replace(tzinfo=timezone.utc))
```

```python
# Change original to match UTC
# Now if we adjust into UTC with astimezone(timezone.utc), 
# we change both the UTC offset and the clock itself.

print(dt.astimezone(timezone.utc))
```
