### Object: Creating Class and Object in Python

```python
# An object (instance) is an instantiation of a class. When class is defined, only the description
# for the object is defined. Therefore, no memory or storage is allocated.
```

```python
# Class
class Student:
    
    # method
    def check_pass_fail(self):
        if self.marks >= 40:
            return True
        else:
            return False

# Creating an Object from Class
# For example, "student1" is an object of class "Student"
student1 = Student()
student1.name = "Ian"
student1.marks = 98

did_pass = student1.check_pass_fail()
print(did_pass)
```

```python
class Car:
    
    # class attribute
    kind = "hybrid"
    
    # instance attribute
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year =  year
        
# instantiate the Car class

black = Car("Tesla", "S", 2021)
red = Car("Toyota", "Camry", 2020)

# access the class attributes
print("Black is a {} car".format(black.__class__.kind))
print("Red is a {} car".format(red.__class__.kind))


# access the instance attribute
print("{}'s model is {} & manufactured in {}".format(black.brand, black.model, black.year))
print("{}'s model is {} & manufactured in {}".format(red.brand, red.model, red.year))
```

### Object: Creating Class and Object in Python

```python
class Parrot:
    
    #class attribute
    species = "bird"
    
    # instance attribute
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
# instantiate the Parrot class
blu = Parrot("Blu", 10)
woo = Parrot("Woo", 15)

# access the class attribute
print("Blu is a {}".format(blu.__class__.species))
print("Woo is also a {}".format(woo.__class__.species))

# Note: We can access the class attribute using __class__.species.

# access the instance attributes
print("{} is {} years old".format( blu.name, blu.age))
print("{} is {} years old".format( woo.name, woo.age))
```

### Methods: Creating Methods in Python

```python
# Methods are functions defined inside the body of a class. They are used to define the behaviors 
# of an object.
```

```python
class Car:
    
    # instance attributes
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year =  year
        
    # instance method
    def feature(self, auto):
        return "{} is a {} car!".format(self.brand, auto)
    
# instantiate the object
black = Car("Tesla", "S", 2021)

# call our instance methods
print(black.feature("'Wonderful'"))

# In the above code, we defined a method called feature(). It is called instance method because
# it is called on an instance object i.e black.
```

### Inheritance

```python
# parent class
class Fancycar:
    
    def __init__(self):
        print("Which car is this?")
    
    def brand(self):
        print("Tesla")
    
    def model1(self):
        print("S")
        

# Child Class
class Outdated(Fancycar):
    
    def __init__(self):
        # call super() function
        super().__init__()
        print("My car?")
        
    def brand(self):
        print("Toyota")
        
    def model2(self):
        print("Runs slower")
        
toyota = Outdated()
toyota.brand()
toyota.model1()

#1) The child class inherits the functions of parent class. For eg. model1() method.

#2) Again, the child class modified the behavior of the parent class. For eg. brand() method. 

#3) Additionally, we use the super() function inside the __init__() method. This allows us to run 
#3) the __init__() method of the parent class inside the child class.
```

### Encapsulation

```python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

c = Computer()
c.sell()

# change the price
c.__maxprice = 1000
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```

### Polymorphism

```python
class Parrot:

    def fly(self):
        print("Parrot can fly")
    
    def swim(self):
        print("Parrot can't swim")

class Penguin:

    def fly(self):
        print("Penguin can't fly")
    
    def swim(self):
        print("Penguin can swim")

# common interface
def flying_test(bird):
    bird.fly()

#instantiate objects
blu = Parrot()
peggy = Penguin()

# passing the object
flying_test(blu)
flying_test(peggy)
```
