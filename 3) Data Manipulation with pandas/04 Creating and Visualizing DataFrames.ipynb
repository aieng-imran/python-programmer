```python
import pandas as pd
import matplotlib.pyplot as plt

homelessness = pd.read_csv('homelessness.csv')
print(homelessness.head())
```

```python
homelessness["individuals"].hist(bins=5)
plt.show()
```

```python
# Bar Plots

homelessness_bar = homelessness.groupby("state")["state_pop"].mean()
print(homelessness_bar.head())
```

```python
# Bar Plots

homelessness_bar.plot(kind="bar", title="State Population")
plt.show()
```

```python
# Line Plots
# Note: You can also slice DataFrames, but first, you need to sort the index.

# Uploading Data
all_stocks_5yr = pd.read_csv('all_stocks_5yr.csv')

# Setting & sorting index
all_stocks_5yr.set_index("date").sort_index()

# Subsetting AAL stock data
AAL  = all_stocks_5yr[all_stocks_5yr["Name"] == "AAL"]

# Plotting
AAL.plot(x="date", y="close", kind="line", rot=45)
plt.legend(["Closing Price"])
plt.show()
```

```python
# Scatter plots

sales_subset = pd.read_csv('sales_subset.csv')
print(sales_subset.head())
```

```python
sales_subset.plot(x="fuel_price_usd_per_l", y = "weekly_sales", kind="scatter")
plt.show()
```

```python
# Layering Plots

homelessness[homelessness["region"] == "Mountain"].hist()
plt.show()
```

```python

homelessness[homelessness["region"] == "Pacific"]["state_pop"].hist(alpha=0.7)
homelessness[homelessness["region"] == "Mountain"]["state_pop"].hist()
plt.show()
```

```python
# Creating Dataframes
# From a list of dictionaries ( constructed row by row)

list_of_dicts = [
    {"name": "Ian", "age":"31", "DOB": "04/25/1990"},
    {"name": "Sam", "age":"41", "DOB": "04/25/1980"}
]

bio = pd.DataFrame(list_of_dicts)
print(bio)
```

```python
# Creating Dataframes
# Dictionary of Lists- by column

dict_of_lists = {
    "name": ["Ian", "Kyle", "Scott"],
    "age": ["31", "35", "42"],
    "DOB": ["1990", "1986", "1979"]
}
bio_2 = pd.DataFrame(dict_of_lists)
print(bio_2)
```
