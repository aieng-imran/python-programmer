```python
# Learn Context Manager in Python in a easy way tutorial
# https://www.youtube.com/watch?v=HRBdhg0lT3U

# Why do we need context manager?

# The number of database connections that can be opened at a time is also 
# limited(just like file descriptors). Therefore context managers are helpful 
# in managing connections to the database as there could be chances that 
# the programmer may forget to close the connection.
```

```python
with open('hello.txt', 'w') as file:
    file.write('hello, world!')

file = open('hello.txt',  'w')
try:
    file.write('hello, world')
finally:
    file.close()
```
