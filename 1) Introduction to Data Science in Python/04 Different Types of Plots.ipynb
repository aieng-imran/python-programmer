# Making Plot

1) Scatterplot

2) Barchart

3) Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

diabetes = pd.read_csv('diabetes.csv')
print(diabetes.head())
```

```python
# Alpha adjusts the transparency
plt.scatter(diabetes.Age, diabetes.Glucose, color='blue', marker='o', alpha=0.4)
plt.xlabel('Age (Years)')
plt.ylabel('Glucose')
plt.show()
```

```python
# Making a bar chart

plt.barh(diabetes.Glucose, diabetes.Insulin, diabetes.BloodPressure)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(diabetes.Glucose)
plt.show()
```
