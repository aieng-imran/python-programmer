# From DataFrame to Visualization

```python
# From matplotlib, import pyplot under the alias plt
# from matplotlib import pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

tesla = pd.read_csv('TSLA_2010_20.csv')
print(tesla.head())
```

```python
x_values = tesla['Date']
y_values = tesla['Close']
a_values = tesla['Date']
b_values = tesla['High']
plt.plot(x_values, y_values)
plt.plot(a_values, b_values)
plt.show()
```

```python
# Alternatively

plt.plot(tesla['Date'], tesla['Close'])
plt.plot(tesla['Date'], tesla['High'])
plt.show()
```

```python
# Alternatively

plt.plot(tesla.Date, tesla.Close)
plt.plot(tesla.Date, tesla.High)
plt.show()
```

```python
# Plot lines
plt.plot(tesla.Date, tesla.Close, label="Closing Price", color="green", linewidth=3, marker='o')
plt.plot(tesla.Date, tesla.Low, label="Lowest Price", color="dodgerblue", linewidth=2, marker='x')

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price Progression", fontsize=30)
plt.text(100, 600, "Hey")
# Change the style 
plt.style.use('seaborn')
# Add a legend
plt.legend()
# Display the plot
plt.show()
```
