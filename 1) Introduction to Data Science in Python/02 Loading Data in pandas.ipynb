# What can pandas do for you?

1) Loading tabular data from different sources

2) Search for particular rows or columns

3) Calculate aggregate statistics

4) Combining data from multiple sources

```python
import pandas as pd
import matplotlib.pyplot as plt

tesla = pd.read_csv('TSLA_2010_20.csv')
```

```python
print(tesla.head())

# Note: Each of these column names is a string. 
```

```python
print(tesla.info())
```

```python
# Selecting columns

plt.plot(tesla['Date'], tesla['Close'])
```

```python
# Selecting with brackets and string

# We type the name of the DataFrame, followed by an open square bracket, 
# followed by the column string (in quotation marks), and closed with a second square bracket. 

# Select the column item from credit_records
# Use brackets and string notation

price = tesla['Close']
print(price)
```

```python
# Select the column item from credit_records
# Use dot notation

price = tesla.Close
print(price)
```

```python
# Common Mistakes in Column Selection
```

```python
# Using Logics with Data Frame

tesla[tesla.Close > 500]
```
