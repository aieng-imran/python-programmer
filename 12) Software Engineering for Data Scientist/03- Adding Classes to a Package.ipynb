```python
class Student:
    
    # Defining passing criteria
    def check_pass_fail(self):
        if self.marks >= 40:
            return True
        else:
            return False
    
    # instance attribute
    def __init__(self, name, marks):
        self.name = name
        self.marks =  marks

# instantiate the Student class
student1 = Student("Ian", 98)
student2 = Student("Sam", 38)

did_pass = student1.check_pass_fail()
print(did_pass)
did_pass = student2.check_pass_fail()
print(did_pass)
```

```python
# Inheritance
# Inheritance is a way of creating a new class for using details of an 
# existing class without modifying it.
```

```python
# Multilevel Inheritance
# Parent class
class Animal:
    def eat(self):
        print("I can eat")
        
# Child class
class Dog(Animal):
    def bark(self):
        print("I can bark")
        
# Child
class Cat(Animal):
    def get_grumpy(self):
        print("I am getting grumpy")
        
dog1 = Dog()
dog1.bark()
dog1.eat()
```

```python
cat1 = Cat()
cat1.eat()
```

```python
# Use of Inheritance in Python
# https://www.programiz.com/python-programming/object-oriented-programming


# parent class
class Bird:
    
    def __init__(self):
        print("Bird is ready")
        
    def swim(self):
        print("Swim Faster")

# Child class
class Penguin(Bird):
    
    def __init__(self):
        # call super() function
        super().__init__()
        print("Penguin is ready")
        
    def swim(self):
        print("Swim Slower")
    
    def run(self):
        print("Run faster")

peggy = Penguin()
peggy.swim()
peggy.run()
```

```python
# Note: The child class inherits the functions of parent class. 
# We can see this from the swim() method.
# Again, the child class modified the behavior of the parent class. 
# We can see this from the whoisThis() method.

# Additionally, we use the super() function inside the __init__() method. 
# This allows us to run the __init__() method of the parent class inside 
# the child class.
```

```python
# Parent
class Polygon:
    def __init__(self, sides):
        self.sides = sides
    
    def display_info(self):
        print("A polygon is a two dimensional shape with straight lines")
    
    def get_perimeter(self):
        perimeter = sum(self.sides)
        return perimeter

# Child
class Triangle(Polygon):
    def display_info(self):
        print("A triangle is a polygon with 3 edges")
        super().display_info()

# Sub-Child
class Quadrilateral(Polygon):
    def display_info(self):
        print("A quadrilateral is a polygon with 4 edges")
        
t1 = Triangle([5, 6, 7])
perimeter = t1.get_perimeter()
print("The perimeter is", perimeter)
t1.display_info()

# Note: if you're going to use multiple inheritance, use super.
# https://www.datacamp.com/community/tutorials/super-multiple-inheritance-diamond-problem
```
