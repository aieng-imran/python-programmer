```python
# Using doctest

def square(x):
    """Square the number x
    :param x: number to square
    :return: x squared
    >>> square(3)
    9
    """
    return x ** x

import doctest
doctest.testmod()

# Note: To use doctest we need to import it and run the testmod command.
# If there were no failed tests the output from testmod would be blank.
# doctests are great for smaller examples, but you often can't cover 
# every case you want to test in a docstring. 
```
